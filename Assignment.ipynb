```python
from google.colab import files


# Check the uploaded file(s)
import pandas as pd

# If 'titanic.csv' is uploaded, you can read it using pandas
df = pd.read_csv('titanic.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
#Q1  Create a Scatter plot of Male and Female Survivors.
#Display Passengers’ Age on the X-axis, and Fare on the Y-axis.
#Note: Use color = ‘green’ for male and color = ‘red’ for female.
import matplotlib.pyplot as plt

# Filter the data for survivors
survivors = df[df['Survived'] == 1]

# Plotting the scatter plot
plt.figure(figsize=(10, 6))

# Plotting males in green
plt.scatter(survivors[survivors['Sex'] == 'male']['Age'],
            survivors[survivors['Sex'] == 'male']['Fare'],
            color='green', label='Male Survivors', alpha=0.6)

# Plotting females in red
plt.scatter(survivors[survivors['Sex'] == 'female']['Age'],
            survivors[survivors['Sex'] == 'female']['Fare'],
            color='red', label='Female Survivors', alpha=0.6)

# Adding labels and title
plt.title('Scatter Plot of Male and Female Survivors')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()

# Display the plot
plt.show()
```

```python
#Q2 Create a Scatter plot of the Male, Female passengers who were dead.

import matplotlib.pyplot as plt

# Filter the data for passengers who did not survive
dead_passengers = df[df['Survived'] == 0]

# Plotting the scatter plot
plt.figure(figsize=(10, 6))

# Plotting males in blue
plt.scatter(dead_passengers[dead_passengers['Sex'] == 'male']['Age'],
            dead_passengers[dead_passengers['Sex'] == 'male']['Fare'],
            color='blue', label='Male Dead Passengers', alpha=0.6)

# Plotting females in yellow
plt.scatter(dead_passengers[dead_passengers['Sex'] == 'female']['Age'],
            dead_passengers[dead_passengers['Sex'] == 'female']['Fare'],
            color='yellow', label='Female Dead Passengers', alpha=0.6)

# Adding labels and title
plt.title('Scatter Plot of Male and Female Dead Passengers')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()

# Display the plot
plt.show()
```

```python
#Q3 Create a Scatter plot of the Male, Female passengers who were dead.
# Creating the 'Number of Relatives' column
df['Number of Relatives'] = df['SibSp'] + df['Parch']

# Display the first few rows to verify
print(df[['SibSp', 'Parch', 'Number of Relatives']].head())

import matplotlib.pyplot as plt

# Filter the data for passengers who did not survive
dead_passengers = df[df['Survived'] == 0]

# Plotting the scatter plot
plt.figure(figsize=(10, 6))

plt.scatter(dead_passengers['Number of Relatives'],
            dead_passengers['Fare'],
            color='purple', alpha=0.6)

# Adding labels and title
plt.title('Scatter Plot of Dead Passengers')
plt.xlabel('Number of Relatives')
plt.ylabel('Fare')

# Display the plot
plt.show()

```

```python
#Q4 Create a new column in the pandas dataframe with the name ‘Number of Relatives’. Create a Scatter plot of the passengers who were alive.

df['Number of Relatives'] = df['SibSp'] + df['Parch']

# Display the first few rows to check the new column
print(df[['SibSp', 'Parch', 'Number of Relatives']].head())

import matplotlib.pyplot as plt

# Filter the data for passengers who survived
alive_passengers = df[df['Survived'] == 1]

# Plotting the scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot for alive passengers
plt.scatter(alive_passengers['Number of Relatives'], alive_passengers['Fare'], color='green', alpha=0.6)

# Adding labels and title
plt.title('Scatter Plot of Alive Passengers')
plt.xlabel('Number of Relatives')
plt.ylabel('Fare')

# Display the plot
plt.show()
```
